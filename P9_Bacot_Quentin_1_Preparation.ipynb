Jointure des tables 

```python
df_Global = pd.merge(DispoAlim, Population, on=["Zone","Année"],indicator = True,how = 'outer')
df_Global.head()
```

```python
#On ne garde les données qui sont présentes uniquement dans les 2 tables, 
#un manque des données sur la disponibilité alimentaire rend l'analyse inefficace pour les pays concernés 

df_Global = df_Global[df_Global["_merge"] == "both"]
del df_Global["_merge"]
df_Global.head()
```

```python
df_Global["Élément"].unique()
```

```python
# On va conserver uniquement les données concernant les poulets
df_Global = df_Global[df_Global["Produit"] == "Viande de Volailles"]
df_Global.head()
```

```python
# On va créer une table avec production ,importation, exportation

df_prod = df_Global[df_Global["Élément"] == "Production"]
df_prod =df_prod.drop(columns=["Année","Code Élément","Élément","Code Produit","Produit"])
df_prod
```

```python
df_import = df_Global[df_Global["Élément"] == "Importations - Quantité"]
df_import = df_import.drop(columns=["Année","Code Élément","Élément","Code Produit","Produit","Population en Milliers"])
df_import
```

```python
df_export = df_Global[df_Global["Élément"] == "Exportations - Quantité"]
df_export = df_export.drop(columns=["Année","Code Élément","Élément","Code Produit","Produit","Population en Milliers"])
df_export
```

```python
df_Global = pd.merge(df_prod, df_export,  on=["Zone"], how = 'inner' )
df_Global = pd.merge(df_Global, df_import,  on=["Zone"], how = 'inner' )
df_Global
```

```python
df_Global.nunique()
```

```python
df_Global = df_Global.rename(columns={"Valeur": "Production (KTonnes)","Valeur_x":"Importation (KTonnes)","Valeur_y":"Exportation (KTonnes)"})
df_Global = df_Global.drop(columns=["Unité","Unité_x","Unité_y"])
df_Global = df_Global.fillna(0)
df_Global
```
